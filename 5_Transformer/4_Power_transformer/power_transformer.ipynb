```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
```

```python
from sklearn.model_selection import train_test_split

from sklearn.model_selection import  cross_val_score
from sklearn.metrics import r2_score

from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('concrete_data.csv')
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape,df.shape)
```

# Without applying anything

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python

y_pred = lr.predict(X_test)
```

```python
print("R2 Score",r2_score(y_test,y_pred))
```

### cross val check

```python
lr = LinearRegression()
print("Cross val check",np.mean(cross_val_score(lr,X,y,scoring="r2",cv=5)))
```

# Check all columns distribution

```python
for col in X_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    plt.title(f"{col}'s pdf distribution")
    sns.distplot(X_train[col])
    
    plt.subplot(122)
    plt.title(f"{col}'s normal prob plot")
    stats.probplot(X_train[col],dist='norm',plot=plt)
    plt.show()
```

# Box-cox transform

```python
powerTf = PowerTransformer(method='box-cox')

X_train_transformed = powerTf.fit_transform(X_train + .000001)
X_test_transformed = powerTf.transform(X_test + .000001)
```

```python
lr = LinearRegression()

lr.fit(X_train_transformed,y_train)

y_pred_AfterPowerTF = lr.predict(X_test_transformed)

print("R2 score after power transformer",r2_score(y_test,y_pred_AfterPowerTF)) 
```

## cross val score

```python
powerTf = PowerTransformer(method='box-cox')

X_transformed = powerTf.fit_transform( X + .000001)
```

```python
lr = LinearRegression()

print("Cross val r2 score",np.mean(cross_val_score(lr,X_transformed,y,scoring="r2")))
```

# check before and after distribution

```python
X_train_transformed = pd.DataFrame(X_train_transformed,columns=X_train.columns)
```

```python
for col in X_train_transformed.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    plt.title( f"Before {col}'s pdf distribution")
    sns.distplot(X_train[col])
    
    plt.subplot(122)
    plt.title(f"Before {col}'s pdf distribution")
    sns.distplot(X_train_transformed[col])
    plt.show()
```

# Yeo Johnson transform

```python
powerTfYJ = PowerTransformer(method='yeo-johnson')

X_train_transformed = powerTfYJ.fit_transform(X_train + .000001)
X_test_transformed = powerTfYJ.transform(X_test + .000001)
```

```python
lr = LinearRegression()

lr.fit(X_train_transformed, y_train)

y_pred_AfterPowerTF = lr.predict(X_test_transformed)

print("R2 score after power transformer", r2_score(y_test, y_pred_AfterPowerTF)) 
```

### cross val score

```python
powerTfYJ = PowerTransformer(method='box-cox')

X_transformed = powerTfYJ.fit_transform(X + .000001)
```

```python
lr = LinearRegression()

print("Cross val r2 score", np.mean(cross_val_score(lr, X_transformed, y, scoring="r2")))
```
