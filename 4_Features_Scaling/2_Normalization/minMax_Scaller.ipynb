```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv',usecols=[0,1,2],names=['Class label', 'Alcohol', 'Malic acid'])
df.sample(10)
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['Class label'],palette={1:'red',2:'green',3:'blue'})
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size=.3,random_state=0)
```

```python
print(X_train.shape,
X_test.shape,
y_train.shape,
y_test.shape,
df.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

mmScaler = MinMaxScaler()
```

```python
X_train_standard = pd.DataFrame(mmScaler.fit_transform(X_train),columns=X_train.columns)
X_test_standard = pd.DataFrame(mmScaler.transform(X_test),columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Alcohol'], X_train['Malic acid'],c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_standard['Alcohol'], X_train_standard['Malic acid'],c=y_train)
ax2.set_title("After Scaling")
plt.show()
```
