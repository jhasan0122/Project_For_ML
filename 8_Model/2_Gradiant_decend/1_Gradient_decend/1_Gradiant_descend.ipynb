# Lets apply Gradient Descent assuming slope is constant m = 78.35 and b = 100

```python
plt.scatter(X,y)
```

```python
m = 78.5
b = 100
y_pred = ((m * X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red",label="OLS")
plt.plot(X,y_pred,color='green',label='b = 0')
plt.legend()
plt.show()
```

## iteration1

```python
m = 78.35
b = 100

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

#### Lets take learning rate = 0.1

```python
lr = 0.1
step_size = loss_slope * lr
step_size
```

####  Calculating the new intercept

```python
b = b - step_size
b
```

```python
y_pred_newB = ((m*X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red",label="OLS")
plt.plot(X,y_pred,color='green',label='b = 0')
plt.plot(X,y_pred_newB,color="purple",label='b = {}'.format(b))
plt.legend()
plt.show()
```

## iteration2

```python
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
print(loss_slope)

step_size = loss_slope * lr
print(step_size)

#  Calculating the new intercept
b2 = b - step_size
print(b2)

y_pred_newB2 = ((m * X) + b2).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color="red", label="OLS")
plt.plot(X, y_pred, color='green', label='b = 0')
plt.plot(X, y_pred_newB, color="purple", label='b = {}'.format(b))
plt.plot(X, y_pred_newB2, color="orange", label='b = {}'.format(b2))
plt.legend()
plt.show()
```

## iteration3

```python
loss_slope = -2 * np.sum(y - m * X.ravel() - b2)
print(loss_slope)

step_size = loss_slope * lr
print(step_size)

#  Calculating the new intercept
b3 = b2 - step_size
print(b3)

y_pred_newB3 = ((m * X) + b3).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color="red", label="OLS")
plt.plot(X, y_pred, color='green', label='b = 0')
plt.plot(X, y_pred_newB, color="purple", label='b = {}'.format(b))
plt.plot(X, y_pred_newB2, color="orange", label='b = {}'.format(b2))
plt.plot(X, y_pred_newB3, color="#B26700", label='b = {}'.format(b3))
plt.legend()
plt.show()
```

## Done with Loop

```python
def drawGraph(b, m ,lr, epoc):
    for i in range(epoc):
        loss_slope = -2 * np.sum(y - m*X.ravel() - b)
        b -= (lr* loss_slope)
        
        y_pred = m*X + b
        
        plt.plot(X,y_pred)
    plt.scatter(X,y);
```

```python
drawGraph(-100,78.35,0.1,10)
```
